```python
#Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#TASK 2:Load The Dataset
import pandas as pd
df=pd.read_csv('/House Price India.csv')
df
df.head()
```

```python
#TASK_3 Univariate Analysis:
sns.countplot(df['number of floors'])
df['number of floors'].value_counts().plot(kind='bar')
```

```python
df['number of views'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
plt.hist(df['Built Year'], bins=10)
plt.xlabel('YEAR')
plt.ylabel('No of houses built')
plt.title('Histogram')
plt.show()
```

```python
sns.displot(df['Built Year'], kde=True)
plt.xlabel('YEAR BUILT IN')
plt.ylabel('NO OF HOUSES BUILT')
plt.title('Distplot')
plt.show()
```

```python
sns.boxplot(x='Built Year', data=df)
plt.xlabel('X-axis label')
plt.title('Boxplot')
plt.show()
```

```python
df.head()
```

```python
#Bi- Variate Analysis
#boxplot
sns.boxplot(x='Number of schools nearby',y='Price',data=df)
plt.show()
```

```python
#barplot
sns.barplot(x='number of bedrooms',y='Price',data=df)
plt.show()
```

```python
#stripplot
sns.stripplot(x='number of floors',y='Price',data=df)
plt.show()
```

```python
#distplot
sns.displot(df, x='condition of the house', y='living area')
plt.show()
```

```python
#Multivariate Analysis
sns.heatmap(pd.crosstab(df['number of floors'],df['grade of the house']))
```

```python
(df.groupby('number of floors').mean()['Price']).plot(kind='bar')
```

```python
#clustermap
sns.clustermap(pd.crosstab(df['waterfront present'],df['number of floors']))
```

```python
#clustermap
sns.clustermap(pd.crosstab(df['condition of the house'],df['number of floors']))
```
